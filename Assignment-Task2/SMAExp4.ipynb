```python
#4) Experiments
import pandas as pd
import re
df = pd.read_csv("/content/tweets_dataset.csv")
def extract_hashtags(text):
    """
    Extracts hashtags from a given text.

    Args:
        text (str): The tweet text.

    Returns:
        list: A list of hashtags found in the text.
    """
    if pd.isnull(text):
        return []

    hashtags = re.findall(r"#\w+", text)
    return hashtags
df["hashtags"] = df["tweet"].apply(extract_hashtags)
df.to_csv("extracted_hashtags.csv", index=False)

print(df.head())
```
