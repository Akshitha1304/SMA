```python
#13)
# Install necessary library (uncomment if not installed)
!pip install vaderSentiment scipy pandas # Install the required package

import pandas as pd
from scipy.stats import pearsonr, spearmanr
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load the dataset
df = pd.read_csv("/content/correlation_analysis_dataset.csv")  # Ensure the dataset is in the same directory

# Initialize VADER Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment score
df["sentiment"] = df["comment"].apply(lambda x: analyzer.polarity_scores(str(x))["compound"])

# Compute Pearson correlation
pearson_corr_likes, _ = pearsonr(df["sentiment"], df["likes"])
pearson_corr_shares, _ = pearsonr(df["sentiment"], df["shares"])

# Compute Spearman correlation
spearman_corr_likes, _ = spearmanr(df["sentiment"], df["likes"])
spearman_corr_shares, _ = spearmanr(df["sentiment"], df["shares"])

# Display correlation results
print(f"Pearson Correlation (Sentiment vs Likes): {pearson_corr_likes:.2f}")
print(f"Pearson Correlation (Sentiment vs Shares): {pearson_corr_shares:.2f}")
print(f"Spearman Correlation (Sentiment vs Likes): {spearman_corr_likes:.2f}")
print(f"Spearman Correlation (Sentiment vs Shares): {spearman_corr_shares:.2f}")

# Save results to a new CSV file
df.to_csv("correlation_analysis_results.csv", index=False)
```
