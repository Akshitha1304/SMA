```python
#6)
import pandas as pd

def calculate_engagement_rate(file_path):
    """
    Calculates the engagement rate for each post in a dataset.

    Parameters:
    file_path (str): Path to the CSV file containing likes, shares, comments, and followers.

    Returns:
    pandas.DataFrame: Original dataframe with an added 'engagement_rate' column.
    """
    # Load dataset
    df = pd.read_csv(file_path)

    # Ensure numeric columns
    df[['likes', 'shares', 'comments', 'followers']] = df[['likes', 'shares', 'comments', 'followers']].fillna(0).astype(int)

    # Calculate engagement rate (avoid division by zero)
    df["engagement_rate"] = ((df["likes"] + df["shares"] + df["comments"]) / df["followers"]) * 100
    df["engagement_rate"] = df["engagement_rate"].round(2)  # Round to 2 decimal places

    return df

# Example usage:
file_path = "/content/Exp6.csv"  # Update with your actual file name
result_df = calculate_engagement_rate(file_path)
print(result_df)
```
