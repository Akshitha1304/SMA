```python
#8)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_demographics(file_path):
    """
    Analyzes user demographics based on age and location.

    Parameters:
    file_path (str): Path to the CSV file containing user profiles.

    Returns:
    None (Displays visualizations).
    """
    # Load dataset
    df = pd.read_csv(file_path)

    # Ensure age column is numeric
    df["age"] = df["age"].astype(int)

    # --- Age Distribution ---
    plt.figure(figsize=(8, 5))
    sns.histplot(df["age"], bins=10, kde=True, color="blue")
    plt.xlabel("Age")
    plt.ylabel("Number of Users")
    plt.title("Age Distribution of Users")
    plt.show()

    # --- Location Distribution ---
    plt.figure(figsize=(8, 5))
    sns.countplot(y=df["location"], order=df["location"].value_counts().index, palette="viridis")
    plt.xlabel("Number of Users")
    plt.ylabel("Location")
    plt.title("User Distribution by Location")
    plt.show()

    # --- Summary Statistics ---
    print("Age Statistics:\n", df["age"].describe())
    print("\nTop Locations:\n", df["location"].value_counts())

# Example usage:
file_path = "/content/8.csv"  # Update with your actual file name
analyze_demographics(file_path)
```
