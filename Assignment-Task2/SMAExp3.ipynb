```python
#3) Experiments: our task is to assign sentiment labels
import nltk
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
nltk.download('vader_lexicon')
df = pd.read_csv("/content/sentiment_analysis_dataset.csv")
sia = SentimentIntensityAnalyzer()
def get_textblob_sentiment(post):
    blob = TextBlob(post)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'
def get_vader_sentiment(post):
    score = sia.polarity_scores(post)
    if score['compound'] > 0.05:
        return 'positive'
    elif score['compound'] < -0.05:
        return 'negative'
    else:
        return 'neutral'
df['TextBlob_Sentiment'] = df['Post'].apply(get_textblob_sentiment)
df['VADER_Sentiment'] = df['Post'].apply(get_vader_sentiment)
df.to_csv("sentiment_analysis_results.csv", index=False)
print(df.head())
```
